```python
train_data_cleaned = train_data.drop(columns=['Cabin', 'Name', 'Ticket'])
test_data_cleaned = test_data.drop(columns=['Cabin', 'Name', 'Ticket'])
```

```python
imputer_age = SimpleImputer(strategy='median')
imputer_embarked = SimpleImputer(strategy='most_frequent')
imputer_fare = SimpleImputer(strategy='median')
```

```python
train_data_cleaned['Age'] = imputer_age.fit_transform(train_data_cleaned[['Age']]).ravel()
train_data_cleaned['Embarked'] = imputer_embarked.fit_transform(train_data_cleaned[['Embarked']]).ravel()
test_data_cleaned['Fare'] = imputer_fare.fit_transform(test_data_cleaned[['Fare']]).ravel()
```

```python
encoder_sex = LabelEncoder()
encoder_sex.fit(pd.concat([train_data_cleaned['Sex'], test_data_cleaned['Sex']]).astype(str).unique())
train_data_cleaned['Sex'] = encoder_sex.transform(train_data_cleaned['Sex'].astype(str))
test_data_cleaned['Sex'] = encoder_sex.transform(test_data_cleaned['Sex'].astype(str))
```
