```python
df['Profit'] = df['capital_gain'] - df['capital_loss']
```

```python
df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

```python
df['age_hours_interaction'] = df['age'] * df['hours_per_week']
```

```python
!pip install ppscore
from sklearn.ensemble import IsolationForest
import ppscore as pps

iso_forest = IsolationForest(contamination=0.1)
df['outlier'] = iso_forest.fit_predict(df[['age', 'hours_per_week', 'capital_gain']])
```

```python
df_cleaned = df[df['outlier'] == 1]
```

```python
pps_matrix = pps.matrix(df_cleaned)
```

```python
correlation_matrix = df_cleaned.corr()
```

```python
print("PPS Matrix:\n", pps_matrix.head())
print("Correlation Matrix:\n", correlation_matrix)
```

conclusion:from PPS
Age and workclass have a very low PPS score of 0.007. This indicates that age has little predictive power over the workclass.

Age and education have a slightly stronger but still low PPS score of 0.044. this suggests a weak, but somewhat more significant relationship between age and education.

conclusion:from correlation
The Correlation matrix highlights strong linear relationships in a few cases (like age with age-hours interaction and capital gain with capital gain log), suggesting that some features might be redundant or highly dependent on others.
