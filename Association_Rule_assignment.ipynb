```python
df=pd.read_excel("Online retail.xlsx")
```

```python
df_cleaned = df[df.columns[0]].str.split(',', expand=True)
df_encoded = pd.get_dummies(df_cleaned.apply(lambda x: pd.Series(x.dropna().unique()), axis=1))
```

```python
!pip install mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
rules.sort_values(by='lift', ascending=False).head()
```

**Interview Questions**

1.What is lift and why is it important in Association rules?

Ans:Lift measures the strength of a rule compared to random chance.

  Importance: Lift > 1 indicates a strong association; products are bought together more often than expected by chance.

2.What is support and Confidence. How do you calculate them?

Ans:Support: Proportion of transactions containing a specific itemset

confidence: Likelihood that the consequent is bought when the antecedent is bought.

calculation:support(A)=Transaction with A / Total Transaction

confidence(A->B)=Transaction Containing A and B both / transaction with A






3.What are some limitations or challenges of Association rules mining?

Ans:Scalability: Computationally expensive with large datasets.
Irrelevant Rules: May generate many trivial or non-actionable rules.
Sparsity: Difficulty in finding meaningful associations in sparse datasets.
